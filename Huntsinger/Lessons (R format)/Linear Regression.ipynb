## Introduction

## Synopsis

Linear regression is a method to construct models to predict numeric values.  Various versions of linear regression all use optimization techniques to find the best fit line (or plane or hyper-plane) through a set of observations.

**Terms**
* **response, dependent, outcome:** the variable to be predicted
* **explanatory, independent, predictor, input:** the variables used to make the prediction

## Exposition

### Simple Linear Regression

Construct a model to predict the numeric value of an outcome variable given one predictor variable.

#### Data

Consider this dataset and new observation.

```r
data = data.frame(x1=c(1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0), outcome=c( 1.0, 0.5, 3.0, 3.0, 4.0, 5.5, 7.0, 6.0))
new = data.frame(x1=5.0)

row.arrange(data %>% captionx, new %>% captionx)

ggplot(data) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) +
geom_point(aes(x=new$x1, y=seq(0,7,length.out=nrow(data))))
```

#### Model: Best Fit Line

```r
model = lm(outcome~x1, data)

data.frame(coefficient=model$coefficients) %>% caption("model", row.names=TRUE)

ggplot(data) +
geom_abline(aes(intercept=model$coefficients["(Intercept)"], slope=model$coefficients["x1"])) + 
geom_point(aes(x=x1, y=outcome), color=PALETTE[3])
```

#### Criterion for Best Fit Line

The best fit line is determined using the ordinary least squares (OLS) optimization method.  Minimize the sum of the squares of the differences between the line and the actual values.

```r
ggplot(data) +
geom_abline(aes(intercept=model$coefficients["(Intercept)"], slope=model$coefficients["x1"])) + 
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model, data))) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) +
geom_point(aes(x=x1, y=predict(model, data)), shape=8)
```

```r
data$predict = predict(model, data)
data$error = data$outcome - data$predict
data$square_error = data$error^2
data %>% caption

data.frame(sum(data$error), sum(data$square_error))
```

There are many lines that minimize differences.  There are many lines that minimize absolute differences.  There is **exactly one** line that minimizes square of differences.

For example, consider this simple case.

```r
data.simple_case = data.frame(x1=c(3,3,6,6), outcome=c(6,3,6,3))
model.simple_case = lm(outcome~x1, data.simple_case)
ggplot(data.simple_case) + xlim(0,10) + ylim(0,10) + 
geom_abline(aes(intercept=model.simple_case$coefficients["(Intercept)"], slope=model.simple_case$coefficients["x1"])) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model.simple_case, data.simple_case))) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3])
```

#### Prediction

```r
model = lm(outcome~x1, data)
new$outcome.predicted = predict(model, new)

new %>% caption

ggplot(data) +
geom_abline(aes(intercept=model$coefficients["(Intercept)"], slope=model$coefficients["x1"])) + 
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) +
geom_point(aes(x=new$x1, y=new$outcome.predicted), color="gold", size=6) +
geom_point(aes(x=new$x1, y=new$outcome.predicted))
```

### Multiple Linear Regression

Construct a model to predict the numeric value of an outcome variable given several predictor variables.

#### Data

Consider this dataset and new observation.

```r
data = data.frame(x1=c(1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0), x2=c(5.0, 7.0, 2.0, 4.0, 3.0, 1.0, 2.0, 2.5), outcome=c( 1.0, 0.5, 3.0, 3.0, 4.0, 5.5, 7.0, 6.0))
new = data.frame(x1=5.0, x2=5.0)

row.arrange(data %>% captionx, new %>% captionx)

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(data, type="s", col=PALETTE[3], xlim=c(0,10), ylim=c(0,10), zlim=c(0,10))
spheres3d(data.frame(x1=new$x1, x2=new$x2, outcome=seq(0,10,1)), radius=0.35, color="grey50", add=TRUE)
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

#### Model

```r
model = lm(outcome~x1+x2, data)

data.frame(coefficient=model$coefficients) %>% caption("model",row.names=TRUE)

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(data, type="s", col=PALETTE[3], xlim=c(0,10), ylim=c(0,10), zlim=c(0,10))
planes3d(a=model$coefficients["x1"], b=model$coefficients["x2"], c=-1, d=model$coefficients["(Intercept)"], color="grey50", alpha=0.3, add=TRUE)
spheres3d(data.frame(x1=new$x1, x2=new$x2, outcome=seq(0,10,1)), radius=0.35, color="grey50", add=TRUE)
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

<br>

#### Prediction

```r
new$outcome.predicted = predict(model, new)
new %>% caption

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(rbind(data, data.frame(x1=new$x1, x2=new$x2, outcome=new$outcome.predicted)), type="s", col=c(rep(PALETTE[3],nrow(data)),NEW_COLOR), xlim=c(0,10), ylim=c(0,10), zlim=c(0,10))
planes3d(a=model$coefficients["x1"], b=model$coefficients["x2"], c=-1, d=model$coefficients["(Intercept)"], color=NEW_COLOR, alpha=0.3, add=TRUE)
segments3d(data.frame(x1=new$x1, x2=new$x2, outcome=c(0,new$outcome.predicted)), color="grey50", lw=5)
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

```r
grid.arrange(ggplot(data) +
             geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) +
             geom_point(aes(x=new$x1, y=new$outcome.predicted), color="gold", size=6) +
             geom_point(aes(x=new$x1, y=new$outcome.predicted)),

             ggplot(data) +
             geom_point(aes(x=x2, y=outcome), color=PALETTE[3]) +
             geom_point(aes(x=new$x2, y=new$outcome.predicted), color="gold", size=6) +
             geom_point(aes(x=new$x2, y=new$outcome.predicted)),
             
             nrow=1)
```

### Linear Regression with Log Outcome Transformation

Construct a model to predict the numeric value of the log of an outcome variable given predictor variable(s).  This may be useful if the variables are not well linearly correlated. 

#### Data

Consider this dataset and new observation.

```r
data = data.frame(x1=c( 1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0), outcome=c( 7.0, 6.5, 6.0, 2.0, 2.0, 1.0, 1.0, 1.5))
new = data.frame(x1=5.0)

row.arrange(data %>% captionx, new %>% captionx)

ggplot(data) + ylim(-1,10) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) +
geom_point(aes(x=new$x1, seq(0,7,length.out=nrow(data))))
```

#### Model & Prediction without Transformation

```r
model = lm(outcome~x1, data)
new$outcome.predicted = predict(model, new)

row.arrange(data.frame(coefficient=model$coefficients) %>% captionx("model",row.names=TRUE), new %>% captionx)

ggplot(data) + ylim(-1,10) + ylab("outcome") +
geom_abline(aes(intercept=model$coefficients["(Intercept)"], slope=model$coefficients["x1"])) + 
geom_point(aes(x=x1, y=predict(model, data)), shape=8) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model, data))) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) +
geom_point(aes(x=new$x1, y=new$outcome.predicted), color="gold", size=6) +
geom_point(aes(x=new$x1, y=new$outcome.predicted))
```

#### Transformation

```r
data$log_outcome = log(data$outcome)
data = data[,c(1,3,2)]
new$outcome.predicted = NULL

row.arrange(data %>% captionx, new %>% captionx)
```

#### Model & Predicition with Transformation

```r
model = lm(log_outcome~x1, data)
new$log_outcome.predicted = predict(model, new)

row.arrange(data.frame(coefficient=model$coefficients) %>% captionx("model", row.names=TRUE), new %>% captionx)

ggplot(data) + ylim(-1,10) + ylab("log(outcome)") +
geom_abline(aes(intercept=model$coefficients["(Intercept)"], slope=model$coefficients["x1"])) + 
geom_point(aes(x=x1, y=predict(model, data)), shape=8) +
geom_segment(aes(x=x1, y=log_outcome, xend=x1, yend=predict(model, data)), color=PALETTE[4]) +
geom_point(aes(x=x1, y=log_outcome), color=PALETTE[3]) +
geom_point(aes(x=new$x1, y=new$log_outcome.predicted), color="gold", size=6) +
geom_point(aes(x=new$x1, y=new$log_outcome.predicted))
```

#### Prediction

Unwind transformation. 

```r
new$outcome.predicted = exp(predict(model, new))

new %>% caption

ggplot(data) + ylim(-1,10) + ylab("outcome") +
geom_smooth(aes(x=x1, y=exp(predict(model, data))), method="loess", se=FALSE) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=exp(predict(model, data)))) +
geom_point(aes(x=x1, y=exp(predict(model, data))), shape=8) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) +
geom_point(aes(x=new$x1, y=new$outcome.predicted), color="gold", size=6) +
geom_point(aes(x=new$x1, y=new$outcome.predicted))
```

### Linear Regression with Polynomial Transformation

Construct a model to predict the numeric value of an outcome variable given predictor variables, where some of the variables are powers of other variables.  This may be useful if the variables are not well linearly correlated. 

#### Data

Consider this dataset and new observation.

```r
data = data.frame(x1=c( 1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0), outcome=c( 8.0, 6.5, 6.0, 2.0, 1.0, 3.0, 5.0, 9.0))
new = data.frame(x1=5.0)

row.arrange(data %>% captionx, new %>% captionx)

ggplot(data) + ylim(0,10) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) +
geom_point(aes(x=new$x1, seq(0,10,length.out=nrow(data))))
```

#### Model & Prediction without Transformation

```r
model = lm(outcome~x1, data)
new$outcome.predicted = predict(model, new)

row.arrange(data.frame(coefficient=model$coefficients) %>% captionx("model", row.names=TRUE), new %>% captionx)

ggplot(data) +
geom_abline(aes(intercept=model$coefficients["(Intercept)"], slope=model$coefficients["x1"])) + 
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model, data))) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) +
geom_point(aes(x=new$x1, y=new$outcome.predicted), color="gold", size=6) +
geom_point(aes(x=new$x1, y=new$outcome.predicted)) +
geom_point(aes(x=x1, y=predict(model, data)), shape=8)
```

#### Transformation

```r
data$x1.square = data$x1^2
data = data[,c(1,3,2)]

new$outcome = NULL
new$x1.square = new$x1^2

row.arrange(data %>% captionx, new[,c(1,3,2)] %>% captionx("new"))
```

#### Model & Prediction with Transformation

Model: $outcome = 12.70 + 0.41{x_1}^2 + -4.14x_1$

```r
model = lm(outcome~x1.square+x1, data)
new$outcome.predicted = predict(model, new)

row.arrange(data.frame(coefficient=model$coefficients) %>% captionx("model", row.names=TRUE), new[,c(1,3,2)] %>% captionx("new"))

ggplot(data) +
geom_smooth(aes(x=x1, y=predict(model, data)), method="loess", se=FALSE) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model, data))) +
geom_point(aes(x=x1, y=predict(model, data)), shape=8) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) +
geom_point(aes(x=new$x1, y=new$outcome.predicted), color="gold", size=6) +
geom_point(aes(x=new$x1, y=new$outcome.predicted)) +
ylab("outcome")
```

#### Easy R Syntax for Linear Regression with Polynomial Transformation

Model: $outcome = 12.70 + 0.41{x_1}^2 + -4.14x_1$

```r
model = lm(outcome~poly(x1,2,raw=TRUE), data)
new$x1.square = NULL
new$outcome.predicted = predict(model, new)

row.arrange(data.frame(coefficient=model$coefficients) %>% captionx("model", row.names=TRUE), new %>% captionx)

ggplot(data) +
geom_smooth(aes(x=x1, y=predict(model, data)), method="loess", se=FALSE) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model, data))) +
geom_point(aes(x=x1, y=predict(model, data)), shape=8) +
geom_point(aes(x=x1, y=outcome), color=PALETTE[3]) +
geom_point(aes(x=new$x1, y=new$outcome.predicted), color="gold", size=6) +
geom_point(aes(x=new$x1, y=new$outcome.predicted)) +
ylab("outcome")
```

### Linear Regression with Interaction Variables

Construct a model to predict the numeric value of an outcome variable given predictor variables and interaction variables.  This may be useful if the variables are not well linearly correlated. 

#### Data

Consider this dataset and new observation.

```r
data = data.frame(x1=c(1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0), x2=c(5.0, 7.0, 2.0, 4.0, 3.0, 1.0, 2.0, 2.5), outcome=c( 1.0, 0.5, 3.0, 3.0, 4.0, 5.5, 7.0, 6.0))
new = data.frame(x1=5.0, x2=5.0)

row.arrange(data %>% captionx, new %>% captionx)

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(data, type="s", col=PALETTE[3], xlim=c(0,10), ylim=c(0,10), zlim=c(0,10))
spheres3d(data.frame(x1=new$x1, x2=new$x2, outcome=seq(0,10,1)), radius=0.35, color=NEW_COLOR, add=TRUE)
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

#### Model & Prediction with Transformation

```r
data$x1x2 = data$x1 * data$x2
new$x1x2 = new$x1 * new$x2

model = lm(outcome~x1+x2+x1x2, data)
new$outcome.predicted = predict(model, new)

row.arrange(data.frame(coefficient=model$coefficients) %>% captionx("model", row.names=TRUE), new %>% captionx)

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(rbind(data, data.frame(x1=new$x1, x2=new$x2, x1x2=new$x1x2, outcome=new$outcome.predicted)), type="s", col=c(rep(PALETTE[3],nrow(data)),"grey50"), xlim=c(0,10), ylim=c(0,10), zlim=c(0,10))
segments3d(data.frame(x1=new$x1, x2=new$x2, outcome=c(0,new$outcome)), color="grey50", lw=5)
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

#### Compare to Prediction by Model without Interaction Variable

```r
model = lm(outcome~x1+x2, data)
new$outcome.predicted = predict(model, new)

data$x1x2 = NULL
new$x1x2 = NULL

row.arrange(data.frame(coefficient=model$coefficients) %>% captionx("model", row.names=TRUE), new %>% captionx)

p = open3d(); view3d(-20, -4, zoom=0.65)
plot3d(rbind(data, data.frame(x1=new$x1, x2=new$x2, outcome=new$outcome.predicted)), type="s", col=c(rep(PALETTE[3],nrow(data)),NEW_COLOR), xlim=c(0,10), ylim=c(0,10), zlim=c(0,10))
segments3d(data.frame(x1=new$x1, x2=new$x2, outcome=c(0,new$outcome)), color=NEW_COLOR, lw=5)
grid3d(c("x","y","z"))
rglwidget(elementId="myplot", width=400, height=400)
```

### Linear Regression with Dummy-Coded Categorical Predictor Variables

The linear regression method assumes all variables are numerical, so convert categorical variables to dummy variables if necessary. 

#### Data

Consider this dataset and new observation.

```r
x2_labels = c("A","B","C")

data = data.frame(x1=c(1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0),
                  x2=factor(c("A","B","C","A","A","B","C","C"), x2_labels),
                  outcome=c( 1.0, 0.5, 3.0, 3.0, 4.0, 5.5, 7.0, 6.0))
new = data.frame(x1=5.0, x2=factor("B", x2_labels))

row.arrange(data %>% captionx, new %>% captionx)

ggplot(data) + geom_point(aes(x=x1, y=outcome, color=x2)) +
geom_point(aes(x=new$x1, y=seq(0,7,length.out=nrow(data))), color=NEW_COLOR, size=6) +
geom_point(aes(x=new$x1, y=seq(0,7,length.out=nrow(data))), color=PALETTE[2]) +
scale_color_manual(values=PALETTE[c(1,2,5)])
```

#### Transformation

Convert categorical variable(s) to dummy variables. 

```r
data.d = dummify(data, drop=FALSE, omit.constants=FALSE)
data.d$x2A = NULL

new.d = dummify(new, drop=FALSE, omit.constants=FALSE)
new.d$x2A = NULL

row.arrange(data.d %>% captionx, new.d %>% captionx)
```

#### Model & Prediction with Transformation

Build a linear regression model based on the data expressed in terms of dummy variables.

Model: $outcome = 0.05 + 0.71x_1 + -0.08x_{2B} + 0.76x_{2C}$

```r
model.d = lm(outcome~., data.d)
new$outcome.predicted = predict(model.d, new.d)

row.arrange(data.frame(coefficient=model.d$coefficients) %>% captionx("model", row.names=TRUE), new %>% captionx)

ggplot(data) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model.d, data.d))) +
geom_point(aes(x=x1, y=outcome, color=x2)) +
geom_point(aes(x=x1, y=predict(model.d, data.d)), shape=8) +
geom_point(aes(x=new$x1, y=new$outcome.predicted), color="gold", size=6) +
geom_point(aes(x=new$x1, y=new$outcome.predicted, color=new$x2)) +
ylab("outcome") +
scale_color_manual(values=PALETTE[c(1,2,5)])
```

### Dummy-Coded vs. Index-Coded 

The two models make different predictions.  The dummies version predicts the existing data slightly better. 

#### Data

Consider this dataset and new observation.

```r
x2_labels = c("A","B","C")

data = data.frame(x1=c(1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0),
                  x2=factor(c("A","B","C","A","A","B","C","C"), x2_labels),
                  outcome=c( 1.0, 0.5, 3.0, 3.0, 4.0, 5.5, 7.0, 6.0))
new = data.frame(x1=5.0, x2=factor("B", x2_labels))

row.arrange(data %>% captionx, new %>% captionx)
```

#### Models

```r
model.dummy = lm(outcome~x1+x2, data) # dummy-coded
model.index = lm(outcome~x1+as.numeric(x2), data) # index-coded

row.arrange(data.frame(coefficient=model.dummy$coefficients) %>% captionx("dummy model", row.names=TRUE),
            data.frame(coefficient=model.index$coefficients) %>% captionx("index model", row.names=TRUE))
```

#### Predictions

```r
data.frame(outcome_dummy=predict(model.dummy, new), outcome_index=predict(model.index, new))

grid.arrange(ggplot(data) +
             geom_point(aes(x=x1, y=predict(model.dummy, data)), shape=8) +
             geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model.dummy, data))) +
             geom_point(aes(x=x1, y=outcome, color=x2)) +
             geom_point(aes(x=new$x1, y=predict(model.dummy, new)), color="gold", size=6) +
             geom_point(aes(x=new$x1, y=predict(model.dummy, new), color=new$x2)) +
             scale_color_manual(values=PALETTE[c(1,2,5)]) +
             ggtitle("Dummy-Coded") + ylab("outcome"),

             ggplot(data) +
             geom_point(aes(x=x1, y=predict(model.index, data)), shape=8) +
             geom_segment(aes(x=x1, y=outcome, xend=x1, yend=predict(model.index, data))) +
             geom_point(aes(x=x1, y=outcome, color=x2)) +
             geom_point(aes(x=new$x1, y=predict(model.index, new)), color="gold", size=6) +
             geom_point(aes(x=new$x1, y=predict(model.index, new), color=new$x2)) +
             scale_color_manual(values=PALETTE[c(1,2,5)]) +
             ggtitle("Index-Coded") + ylab("outcome"),

             nrow=1)
```

### Example of Complex Linear Regression

You can combine and expand log and polynomial linear regression to form more complex models.

#### Data

Consider this dataset and new observation.

```r
x2_labels = c("A","B","C")
data = data.frame(x1=c(1.0, 2.0, 3.0, 4.0, 6.0, 6.5, 7.0, 9.0), 
                  x2=factor(c("A","B","C","A","B","C","A","A"), x2_labels),
                  outcome=c( 1.0, 3.5, 6.0, 2.0, 1.0, 3.0, 5.0, 9.0))
new = data.frame(x1=5.0, x2=factor("A", x2_labels))

row.arrange(data %>% captionx, new %>% captionx)

ggplot(data) + geom_point(aes(x=x1, y=outcome, color=x2)) +
geom_point(aes(x=new$x1, y=seq(0,10,length.out=nrow(data))), color=NEW_COLOR, size=6) +
geom_point(aes(x=new$x1, y=seq(0,10,length.out=nrow(data))), color=PALETTE[1]) +
scale_color_manual(values=PALETTE[c(1,2,5)]) + ylim(0,10)
```

#### Model & Prediction with Transformation

Model: $outcome = 15.31 + -0.083{x_1}^3 + 1.94{x_1}^2 + -17.18x_1 + 20.57log(x_1) + -0.97x_{2B} + 0.07x_{2C}$

```r
model = lm(log(outcome) ~ poly(x1,3,raw=TRUE) + log(x1) + x2, data)
new$outcome.predicted = predict(model, new)

row.arrange(data.frame(coefficient=model$coefficients) %>% captionx("model", row.names=TRUE), new %>% captionx)

ggplot(data) +
geom_segment(aes(x=x1, y=outcome, xend=x1, yend=exp(predict(model, data)))) +
geom_point(aes(x=x1, y=outcome, color=x2)) +
geom_point(aes(x=new$x1, y=exp(new$outcome.predicted)), color="gold", size=6) +
geom_point(aes(x=new$x1, y=exp(new$outcome.predicted), color=new$x2)) +
geom_point(aes(x=x1, y=exp(predict(model, data))), shape=8)
```
